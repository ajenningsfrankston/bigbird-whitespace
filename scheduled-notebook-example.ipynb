```python
# From https://www.kite.com/python/answers/how-to-download-a-csv-file-from-a-url-in-python

import requests

csv_url = "https://www1.ncdc.noaa.gov/pub/data/cdo/samples/PRECIP_HLY_sample_csv.csv"

req = requests.get(csv_url)
url_content = req.content
csv_file = open('precipitation.csv', 'wb')

csv_file.write(url_content)
csv_file.close()
```
